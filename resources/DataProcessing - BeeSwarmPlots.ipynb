```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
import numpy as np
filename = "data.csv"
df = pd.read_csv(filename)
df.head()
```

```python
df_Male = df[df["Gender"] == "Male"]
df_Female = df[df["Gender"] == "Female"]

num_bins = 20
bins = np.linspace(min(df_Female["Age"].min(), df_Male["Age"].min()), max(df_Female["Age"].max(), df_Male["Age"].max()), num_bins)
plt.hist([df_Male["Age"], df_Female["Age"]], bins=bins, label="Stacked", alpha=0.5)

plt.legend()
plt.ylabel("Counts")
plt.xlabel("Ages")
```

```python
df_Visual = df[df["Condition"] == "visual"]
df_Written = df[df["Condition"] == "written"]
df_Control = df[df["Condition"] == "control"]
df_All = df[df["Condition"] == "all"]


num_bins = 20
#bins = np.linspace(min(df_Female["Age"].min(), df_Male["Age"].min()), max(df_Female["Age"].max(), df_Male["Age"].max()), num_bins)
plt.hist([df_Visual["Robot interaction"], df_Written["Robot interaction"], df_All["Robot interaction"], df_Control["Robot interaction"]], label=["v","c","w","a"], bins=5, alpha=0.5)

plt.legend()
plt.ylabel("Counts")
plt.xlabel("Ages")
```
