# Plotting
### Different plotting methods

```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
```

```python
gender = {'Male': 1,'Female': 2}
#df.Gender = [gender[item] for item in df.Gender]
df_Male = df[df["Gender"]==1]
df_Female = df[df["Gender"]!=1]
# plt.hist(df_Male, label="Male")
# plt.hist(df_Female, label="Female")
```

```python
plt.hist(df_Male["Age"], label="Male")
plt.hist(df_Female["Age"], label="Female")
plt.legend()
plt.ylabel("Counts")
plt.xlabel("Age groups")
```

```python
bins = np.linspace(min(df_Female["Age"].min(), df_Male["Age"].min()), max(df_Female["Age"].max(), df_Male["Age"].max()), 100)
plt.hist(df_Male["Age"], bins=bins, label="Male", histtype="step", lw=1)
plt.hist(df_Female["Age"], bins = bins, label="Female", histtype="step", lw=2,ls=":")
plt.legend()
plt.ylabel("Counts")
plt.xlabel("Ages")
```

```python
bins = np.linspace(min(df_Female["Age"].min(), df_Male["Age"].min()), max(df_Female["Age"].max(), df_Male["Age"].max()), 100)
plt.hist(df_Male["Age"], bins=bins, label="Male", histtype="step", lw=1)
plt.hist(df_Female["Age"], bins = bins, label="Female", histtype="step", lw=2,ls=":")
plt.hist([df_Male["Age"], df_Female["Age"]], bins=bins, label="Stacked", density=True, histtype="barstacked", alpha=0.5)
plt.hist([df_Male["Age"], df_Female["Age"]], bins=bins, label="Stacked", histtype="barstacked", alpha=0.5)

plt.legend()
plt.ylabel("Counts")
plt.xlabel("Ages")
```

```python
num_bins = 20
bins = np.linspace(min(df_Female["Age"].min(), df_Male["Age"].min()), max(df_Female["Age"].max(), df_Male["Age"].max()), num_bins)
#plt.hist(df_Male["Age"], bins=bins, label="Male", histtype="step", lw=1)
#plt.hist(df_Female["Age"], bins = bins, label="Female", histtype="step", lw=2,ls=":")
plt.hist([df_Male["Age"], df_Female["Age"]], bins=bins, label="Stacked", alpha=0.5)

plt.legend()
plt.ylabel("Counts")
plt.xlabel("Ages")
```
