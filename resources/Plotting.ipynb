```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import gaussian_kde

data = pd.read_csv('data.csv')
#data['LogTime(A)'] = np.log(data['TotalTime(A)'])
#data['LogTime(B)'] = np.log(data['TotalTime(B)'])

# Data for plotting
x = data['NASA overall']
y = data['NASA overall']
a = data['G2 Total']
b = data['G2 Total']
    

densityx = gaussian_kde(x)
xs = np.linspace(2,6,200)
densityx.covariance_factor = lambda : .25
densityx._compute_covariance()


densitya = gaussian_kde(a)
densitya.covariance_factor = lambda : .25
densitya._compute_covariance()

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Sharing x per column, y per row')
ax1.plot(xs,densityx(xs))
ax2.plot(xs,densitya(xs))
ax3.plot(x, 'tab:green')
ax4.plot(x, 'tab:red')

for ax in fig.get_axes():
    ax.label_outer()
```
